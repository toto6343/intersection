```python
import requests
import urllib
import json
import pandas as pd
import cv2
```

```python
# 인증키
key = "API키"

# 도로 유형(ex: 고속도로 / its: 국도)
Type = "its"

# CCTV 유형(1: 실시간 스트리밍 / 2: 동영상 파일 3: 정지 영상)
cctvType = "1"

# 최소 경도 영역
minX = float(120.95)

# 최대 경도 영역
maxX = float(127.02)

# 최소 위도 영역
minY = float(30.55)

# 최대 위도 영역
maxY = float(38.60)

# 출력 결과 형식
getType = "json"
```

```python
# TEST1
# url_cctv = f'https://openapi.its.go.kr:9443/cctvInfo?apiKey={key}&type=ex&cctvType=1&minX=126.756&maxX=126.7936&minY=37.48135&maxY=37.5843&getType=json'

# TEST2
url_cctv = f"https://openapi.its.go.kr:9443/cctvInfo?apiKey={key}&type={Type}&cctvType={cctvType}&minX={minX}&maxX={maxX}&minY={minY}&maxY={maxY}&getType={getType}"
```

```python
response = urllib.request.urlopen(url_cctv)
```

```python
json_str = response.read().decode("utf-8")
```

```python
json_object = json.loads(json_str)
```

```python
cctv_play = pd.json_normalize(json_object["response"]["data"], sep=',')
```

```python
import cv2
import pandas as pd
import matplotlib.pyplot as plt
import os
from tqdm import tqdm

# CCTV URL과 이름 리스트
test_url_list = cctv_play["cctvurl"]
cctv_name_list = cctv_play["cctvname"]

# 저장할 프레임 수
frames_per_cctv = 100

# 전체 CCTV 대상으로 반복
for idx, (url, cctv_name_raw) in enumerate(zip(test_url_list, cctv_name_list)):

    cap = cv2.VideoCapture(url)
    if not cap.isOpened():
        print(f"[{idx}] ❌ Can't open: {url}")
        continue

    # CCTV 이름 정제 → 폴더/파일 이름용
    cctv_name = cctv_name_raw.strip().replace("[", "").replace("]", "").replace(" ", "_")
    folder_name = f"{idx}_{cctv_name}"

    # 폴더 생성
    os.makedirs(folder_name, exist_ok=True)

    print(f"[{idx}] ▶ {folder_name} 에서 프레임 수집 시작...")

    saved_count = 0
    while saved_count < frames_per_cctv:
        ret, frame = cap.read()
        if not ret:
            print(f"[{idx}] ⚠️ 프레임 수집 중단 (총 {saved_count}개 저장됨)")
            break

        # RGB → BGR 변환 (저장용)
        filepath = os.path.join(folder_name, f"{cctv_name}_{saved_count + 1:03}.jpg")
        cv2.imwrite(filepath, frame)

        saved_count += 1

    cap.release()

    print(f"[{idx}] ✅ {saved_count}개의 프레임 저장 완료 → {folder_name}")

print("🎉 전체 프레임 저장 완료")
```
